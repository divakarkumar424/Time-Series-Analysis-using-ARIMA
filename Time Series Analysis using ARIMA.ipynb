### Need To Know About ARIMA

ARIMA or Auto-Regressive Integrated Moving Average is a very widely used statistical method for time series analysis and forecasting. It is a class of models that capture a spectrum of different standard temporal structures present in time series data. By implementing an ARIMA model, you can forecast and analyze a time series using past values, such as predicting future prices based on historical earnings. 

The ARIMA is applied for forecasting with the univariate time series data. This model is applied in some cases of time series forecasting where data shows the evidence of non-stationarity in the sense of mean, where an initial differencing step can be applied one or more times to eliminate the non-stationarity of the mean function.

The ARIMA has three parts in it:-

1. AR (Auto-Regressive): This is a process of regressing a variable on past values of itself which means finding the relationship between current and past lagged values.
2. Integrated: This property reduces the seasonality in the time series and helps in making the time series stationary.
3. Moving Average: It removes non-determinism or random movements from a time series.

The ARIMA model works on the ideal assumption of the data being stationary. A standard notation used for describing ARIMA is by parameters p,d and q. The parameters are substituted with an integer value to indicate the specific ARIMA model being used quickly.

The parameters of the ARIMA model are further described as follows:

* p: Stands for the number of lag observations included in the model, also known as the lag order. This specifies the autoregressive property of the ARIMA model.
* d: The number of times the raw observations are differentiated, also called the degree of differencing. This specifies the integrated property of the ARIMA model.
* q: Is the size of the moving average lags and also called the order of moving average.

##### How to find the values for p, q and d?

Let’s understand the important points that need to be considered when finding the values for p, q and d.

###### To find d:

The purpose of differencing is to make the time series stationary. But we should be careful to not over-difference in the series. An over-differenced series may still be stationary, which in turn will affect the model parameters.

So we should determine the right order of differencing (d). The right order of differencing is the minimum differencing required to get a near-stationary series which moves around a defined mean and the ACF (Autocorrelation) plot reaches zero quickly.

If the autocorrelations are positive for many lags (10 or more), then the series needs further differencing. On the other hand, if the lag 1 autocorrelation itself is too negative, then the series is probably over-differenced.

If we can’t really decide between two orders of differencing, then we go with the order that gives the least standard deviation in the differenced series.

You need differencing only if the series is non-stationary. Else, no differencing is needed, that is, d=0.

###### To find p:

The next step is to identify if the model needs any autoregressive (AR) terms (p). We can find out the required number of AR terms by inspecting the Partial Autocorrelation (PACF) plot.

Partial autocorrelation can be imagined as the correlation between the series and its lag, after excluding the contributions from the intermediate lags. So, PACF indicates the pure correlation between a lag and the series. This way, we can know if that lag is needed in the AR term or not.

Partial autocorrelation of lag (k) of a series is the coefficient of that lag in the autoregression equation of Y.


 

Suppose, if Yt is the current series and Yt-1 is the lag 1 of Y, then the partial autocorrelation of lag 3 (Yt-3) is the coefficient 𝛼3 of Yt-3 in the above equation.

Now, we can find the number of AR terms. Any autocorrelation in a stationarized series can be rectified by adding enough AR terms. So, we initially take the order of the AR term to be equal to as many lags that cross the significance limit in the PACF plot.

###### To find q:

Just like how we looked at the PACF plot for the number of AR terms, we will look at the ACF (Autocorrelation) plot for the number of MA terms (q). An MA term is technically, the error of the lagged forecast.

The ACF tells how many MA terms are required to remove any autocorrelation in the stationarized series.

So, overall, the following important steps are taken when using ARIMA:-

* Find the stationarity in time series with statistical test (ADF) and decide the value of d.
* Find the values for autoregression order (p) and moving average order (q) through ACF and PACF plots.
* Define the ARIMA model using the above values of parameters, fit it and make predictions using it.

###### Step 2: Checking the stationarity of the series

Stationarity means that there should not be any trend component in the series, this can be checked by observing the series pictorially or using a statistical test. First, we will check pictorially.        

```python
# check how series looks like
data.plot()
```

By observing the above plot we can see that the series has a trend component from the beginning but some samples from the year onward 2002 to the end look stationary. Alternatively, we can check this by the Adfuller test.  

The Augmented Dickey-Fuller test is used to measure the stationarity of the series which aims to reject the null hypothesis that the given series is non-stationary. Basically, it calculates the p-value and compares it with a standard threshold value which is 0.05, if the p-value is less than it then the data is stationary.

Now below we will test the original series if it is non-stationary we will calculate the p-value for the 1st differentiation and 2nd differentiation.

```python
# Statistical way of testing the stationarity
from statsmodels.tsa.stattools import adfuller

# H0: Non-Stationary (Accepted if p > 0.05)

#H1: Stationary (Accepted if p < 0.05)

# checking the stationarity
result = adfuller(data['Production_units'])
print('No differentiation: \n P-Value',result[1])

result = adfuller(data['Production_units'].diff().dropna())
print('1st differentiation: \n P-Value',result[1])

result = adfuller(data['Production_units'].diff().diff().dropna())
print('2nd differentiation: \n P-Value',result[1]) 
```

As we can see that the original series is non-stationary as the p-value is greater than 0.05, later the p-value for 1st and 2nd differentiation is smaller than a threshold value. 

If the series has positive autocorrelations out to a high number of lags, then it probably needs a higher order of differencing. The optimal order of differencing is often the order of differencing at which the standard deviation is lowest. So we will choose 2nd differentiation for further analysis. So we will choose 1st differentiation for further analysis and set the value of d as 1.

```python
# Using 1st order differentiation
data['Stationary'] = data['Production_units'].diff()
```

###### Step 3: Calculating the values for p and q

As we discussed in the theoretical section the ARIMA model assumes certain objectives to satisfy those we have to calculate the p: order of the autoregressive model, d: order of differentiation, and q: is the lagged error. 

We already know the d value which is 1, next to calculate p and q we will plot the Autocorrelation and Partial Autocorrelation plot as below.

```python
# import acf and pacf plots to calculate p and q values
import statsmodels.api as sm
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(data['Stationary'].dropna(),lags=40,ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(data['Stationary'].dropna(),lags=40,ax=ax2)
```

From the above ACF plot, we can choose p by observing the most significant lag and its 1st lag. And q to be 10 is calculated based on the number of lags crossing the threshold limits.

* Thus we can consider p = 1, d = 1 (differentiation) and q = 10


##### Step 4: Building the ARIMA model

Now that we have identified the orders of parameters that need to be added while initializing the ARIMA model, below we will initialize the model. First, we will import ARIMA from the statsmodels library and initialize it and will train by using the fit() method on the original time series, not the differentiated one. 

```python
# Import arima model
from statsmodels.tsa.arima_model import ARIMA
```

```python
z# Initialize and train
arm = ARIMA(data.Production_units.dropna(), order=(1,1,10))
model = arm.fit()
```

After successfully training the model, let’s take a look at the summary of the result.

```python
# Summary of the model
model.summary()
```

Above we can see the model summary which provides several statistical measures by which we can evaluate the model. We can also see their total of 11 parameters generated 1 for autoregressive and 10 for moving average. 

Apart from this, we have scores for AIC, BIC, HQIC and standard deviation of innovation which is the difference between the real value at time t and the predicted value at the same time. 

###### Step 5: Forecasting with ARIMA

In this section first, we will see how the model has fitted real data, for this we will plot the series using the plot_predict method.

```python
# Actual vs Fitted values
model.plot_predict(dynamic=False)
```

Now we will forecast 10 months of future units of Electrical energy production.

```python
# forecasting next 10 values
model.forecast(10)[0]
```

Finally, graphically we will check the actual values and forecasted values.

```python
# comparing the fitted interval and forecasting results
model.plot_predict(start = 320,end = 450, dynamic=False)
```

As we can see the model has identified the trend accurately and the same has been replicated in forecasted values.
